```python

from sklearn.feature_selection import SelectKBest,f_classif
from sklearn.decomposition import PCA

# Split the data into features and target variable
X = data.drop(['id','diagnosis','Unnamed: 32'], axis = 1)
y = data['diagnosis']

# Find the best features using SelectKBest
selector = SelectKBest(f_classif, k=20)
selector.fit(X, y)

# Get the indices of the selected features
indices = selector.get_support()

# Print the names of the selected features
features = X.columns[indices]
print(features)

# Reduce the dimensionality of the data using PCA
pca = PCA(n_components=2)
pca.fit(X)

# Project the data onto the lower-dimensional subspace
X_reduced = pca.transform(X)

# Print the shape of the reduced data
print(X_reduced.shape)
```

```python
X= data[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean','compactness_mean', 'concavity_mean',
         'concave points_mean','radius_se', 'perimeter_se', 'area_se', 'concave points_se','radius_worst', 'texture_worst', 
         'perimeter_worst', 'area_worst','smoothness_worst', 'compactness_worst', 'concavity_worst','concave points_worst', 'symmetry_worst']]
y=y = data['diagnosis']

logModel = LogisticRegression()

param_grid = [    
    {'penalty' : ['l1', 'l2', 'elasticnet', 'none'],
    'C' : np.logspace(-4, 4, 20),
    'solver' : ['lbfgs','newton-cg','liblinear','sag','saga'],
    'max_iter' : [100, 1000]
    }
]
```

```python
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(logModel, param_grid = param_grid, cv = 3, verbose=True, n_jobs=-1)
best_clf = clf.fit(X,y)
```

```python
best_clf.best_estimator_
```

```python
logModel = LogisticRegression(C=206.913808111479, max_iter=1000, penalty='l1',solver='liblinear')
logModel.fit(X,y)
print (f'Accuracy - : {logModel.score(X,y):.3f}')
```
