```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#  LOADING DATASET TO PANDAS DATA FRAME
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

Under Sampling

Build a Sample Data Set containing similar distribution of normal transactions and Fraudental transactions

```python
legit_samples = legit.sample(n=169)
```

```python
new_df = pd.concat([legit_samples,fraud], axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
credit_card_data.groupby('Class').mean()
```

Splitting the data into Features and Target

```python
X= new_df.drop(columns='Class', axis=1)
y= new_df['Class']
```

```python
print(X)
```

```python
print(y)
```

Split the data into Training and testing data

```python
X_train , X_test , y_train , y_test = train_test_split(X, y ,test_size=0.2, stratify=y,random_state=2)
```

```python
print(X_train.shape , X_test.shape,y_train.shape,y_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
#  Training the logistic regression moedl with training data

model.fit(X_train,y_train)
```

Model Evaluation 

  Accuracy Score



```python
# Accuracy on training data
X_train_prediction= model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,y_train)
print('Accuracy on training data: ', training_data_accuracy)
```

```python
# Accuracy on testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction , y_test)
print('Accuracy score on testing data : ', testing_data_accuracy)
```
